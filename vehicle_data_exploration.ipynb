```python
# Import pandas and load Auto.csv data.
import pandas as pd

auto = pd.read_csv("../data/Auto.csv", na_values = ['?'])
```

```python
# Inspect Auto.csv data
auto.info()
```

```python
# Inspect Auto.csv data
auto.describe().T
```

```python
# Inspect horsepower variable
auto['horsepower'].value_counts().sort_index()
```

```python
# Import Seaborn Library
import seaborn as sns

# Display correlation matrix for Auto.csv data
corr_matrix_auto = auto.corr(numeric_only = True).round(2)
corr_matrix_auto
```

```python
# Display correlation matrix for Auto.csv data as a heatmap plot
sns.heatmap(data=corr_matrix_auto,
            vmin=-1,
            vmax=1,
            cmap='coolwarm',
            linewidths=.5,
            annot=True)
```

```python
# Create Box Plot for mpg variale
auto['mpg'].plot.box()
```

```python
# Create Scatter plot for variables displacement and mpg
auto.plot.scatter(x='displacement', 
                  y='mpg',
                 grid=True,
                 marker='.',
                 color='purple')
```

```python
# Create a subset for auto data holding cylinders, mpg, year and weight variables.
auto_sub = auto[['cylinders', 'mpg', 'weight', 'year']]

# Filter the auto_sub dataframe to cars made before 1980
auto_intermediate = auto_sub[auto_sub['year'] < 80]

# Re-write the auto_sub data to drop the year column
auto_sub = auto_intermediate.drop(columns=['year'])
auto_sub
```

```python
# Scatter plot for weight versus mpg.
auto_sub.plot.scatter(x='weight',
                 y='mpg',
                grid=True,
                marker='.',
                color='purple')
```

`The scatter plot shows there is a negative correlation between the weight of the vehicle and the miles per gallon, where as the weight increases then the mpg decreases.`
